### Model Development

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = data_analysis.drop(['MEDV','MEDV_log','MEDV_sqrt','MEDV_boxcox'],axis=1)
y = data_analysis['MEDV']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 12345)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test) # Predict MEDV for test observations
```

```python
print(len(y_pred))
print(len(y_test))
```

```python
# Looking at the distrubution of residuals
res = y_test-y_pred
plt.hist(res)
plt.show()
```

```python
# Testing the assumptions
plt.scatter(res,y_pred)
plt.xlabel('Residuals')
plt.ylabel('Predicted Values')
plt.show()
```

```python
#This statistic will always be between 0 and 4. The closer to 0 the statistic, the more evidence for positive serial correlation. The closer to 4, the more evidence for negative serial correlation.
sm.stats.stattools.durbin_watson(res, axis=0) #closer to 2 means there is no autocorrelation
```

```python
# Model intercept
lm.intercept_
```

```python
# Regression coefficients
lm.coef_
```

```python
SSE = sum(np.square(np.array(y_test) - lm.predict(X_test)))
```

```python
RMSE = np.sqrt(SSE) # 57.521
```

```python
Rsq_test = lm.score(X_test,y_test) # 0.69
```

```python
Rsq_train = lm.score(X_train,y_train) # 0.75
```
